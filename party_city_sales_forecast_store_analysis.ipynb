```python
#Party city sales forecast store analysis

#Load Data
import pandas as pd

reported = pd.read_csv("CSV system2_case_study_reported_numbers.xlsx - Reported Numbers.csv")
location = pd.read_csv("CSV system2_case_study_transaction_data_location_info - system2_case_study_transaction_data_location_info.csv")
sales = pd.read_csv("CSV system2_case_study_transaction_data_monthly_sales_by_locationid.csv")
webscrape = pd.read_csv("CSV system2_case_study_webscrape_data_locations.csv")


#Verify Data
print(reported.shape)
print(location.shape)
print(sales.shape)
print(webscrape.shape)

print("\n" + "="*50 + "\n")

print("=== REPORTED NUMBERS ===")
print(reported.head())
print()
print("=== LOCATION INFO ===")
print(location.head())
print()
print("=== MONTHLY SALES ===")
print(sales.head())
print()
print("=== WEBSCRAPE ===")
print(webscrape.head())
```

```python
# reload raw data to start fresh
sales = pd.read_csv("CSV system2_case_study_transaction_data_monthly_sales_by_locationid.csv")
location = pd.read_csv("CSV system2_case_study_transaction_data_location_info - system2_case_study_transaction_data_location_info.csv")

# convert date columns from strings to actual datetime objects
sales['period_start'] = pd.to_datetime(sales['period_start'])
location['est_open_date'] = pd.to_datetime(location['est_open_date'])
location['est_close_date'] = pd.to_datetime(location['est_close_date'])

# assign each month to a quarter
sales['quarter'] = sales['period_start'].dt.to_period('Q')

# bring in open/close dates from location file
sales = sales.merge(location, on='locationid', how='left')

# drop any months where the store wasn't actually open
sales = sales[(sales['period_start'] >= sales['est_open_date']) & 
              (sales['period_start'] <= sales['est_close_date'])]

# sum sales up to the quarter level per store
quarterly = sales.groupby(['locationid', 'quarter'])['amount'].sum().reset_index()

print(quarterly.shape)

print("\n" + "="*50 + "\n")

print(quarterly.head(10))
```

```python
# for each store/quarter combo, look up what that same store did in the same quarter one year prior
quarterly['quarter_dt'] = quarterly['quarter'].dt.to_timestamp()
quarterly['prior_year_quarter'] = quarterly['quarter_dt'] - pd.DateOffset(years=1)
quarterly['prior_year_quarter'] = quarterly['prior_year_quarter'].dt.to_period('Q')

# merge the table onto itself to get current and prior year sales in the same row
same_store = quarterly.merge(
    quarterly[['locationid', 'quarter', 'amount']].rename(columns={'quarter': 'prior_year_quarter', 'amount': 'prior_amount'}),
    on=['locationid', 'prior_year_quarter'],
    how='inner'
)

#only keep stores that have data in BOTH the current and prior year quarter
print(same_store.shape)

print("\n" + "="*50 + "\n")

print(same_store.head(10))
```

```python
# calculate yoy % change for each store
same_store['yoy_pct'] = (same_store['amount'] - same_store['prior_amount']) / same_store['prior_amount'] * 100

# average across all stores to get one number per quarter
our_estimate = same_store.groupby('quarter')['yoy_pct'].mean().reset_index()
our_estimate.columns = ['quarter', 'our_yoy']

print(our_estimate.shape)

print("\n" + "="*50 + "\n")

print(our_estimate)
```

```python
# remove any rows where prior year sales were zero
same_store = same_store[same_store['prior_amount'] > 0]

# recalculate yoy % change for each store
same_store['yoy_pct'] = (same_store['amount'] - same_store['prior_amount']) / same_store['prior_amount'] * 100

# average across all stores per quarter
our_estimate = same_store.groupby('quarter')['yoy_pct'].mean().reset_index()
our_estimate.columns = ['quarter', 'our_yoy']

print(our_estimate.shape)

print("\n" + "="*50 + "\n")

print(our_estimate)
```

```python
# instead of averaging store-level % changes, sum all stores first then calculate yoy
total_by_quarter = same_store.groupby('quarter')[['amount', 'prior_amount']].sum().reset_index()

# now calculate yoy on the totals
total_by_quarter['our_yoy'] = (total_by_quarter['amount'] - total_by_quarter['prior_amount']) / total_by_quarter['prior_amount'] * 100

print(total_by_quarter.shape)

print("\n" + "="*50 + "\n")

print(total_by_quarter)
```

```python
# clean up the reported numbers so we can compare
reported['period_end_dt'] = pd.to_datetime(reported['period_end_dt'])
reported['quarter'] = reported['period_end_dt'].dt.to_period('Q')
reported['reported_yoy'] = reported['reported_yoy'].str.replace('%', '').astype(float)

# merge our estimates with the official reported numbers
comparison = total_by_quarter.merge(reported[['quarter', 'reported_yoy']], on='quarter', how='left')

print(comparison[['quarter', 'our_yoy', 'reported_yoy']])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(comparison['quarter'].astype(str), comparison['our_yoy'], marker='o', label='Our Estimate', color='steelblue')
ax.plot(comparison['quarter'].astype(str), comparison['reported_yoy'], marker='o', label='Party City Reported', color='darkorange')

# vertical line showing where the prediction starts
ax.axvline(x='2022Q4', color='red', linestyle='--', linewidth=1, label='Forecast Start')

ax.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
ax.set_xlabel('Quarter')
ax.set_ylabel('YoY % Change')
ax.set_title('Our Estimate vs Party City Reported Same Store Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# drop quarters where we don't have both numbers
regression_data = comparison.dropna(subset=['our_yoy', 'reported_yoy'])

x = regression_data['our_yoy'].values
y = regression_data['reported_yoy'].values

# run the regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# generate the fitted line
x_line = np.linspace(x.min(), x.max(), 100)
y_line = slope * x_line + intercept

# plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(x, y, color='steelblue', zorder=5, label='Quarters')
ax.plot(x_line, y_line, color='red', linewidth=1.5, label=f'y = {slope:.3f}x + {intercept:.3f}')

# annotate each point with the quarter label
for i, row in regression_data.iterrows():
    ax.annotate(str(row['quarter']), (row['our_yoy'], row['reported_yoy']),
                textcoords='offset points', xytext=(5, 5), fontsize=8)

ax.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
ax.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
ax.set_xlabel('Our Estimate (YoY %)')
ax.set_ylabel('Party City Reported (YoY %)')
ax.set_title('OLS Regression: Transaction Estimate vs Reported Same Store Sales')
ax.legend()

# print stats
print(f"R²:        {r_value**2:.4f}")
print(f"Slope:     {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"P-value:   {p_value:.6f}")
print(f"Std Error: {std_err:.4f}")
print(f"N:         {len(x)}")

plt.tight_layout()
plt.show()
```

```python
# grab our raw q4 2022 estimate
q4_2022_raw = total_by_quarter[total_by_quarter['quarter'] == '2022Q4']['our_yoy'].values[0]

# plug it into the regression equation to get our adjusted prediction
q4_2022_predicted = slope * q4_2022_raw + intercept

print(f"Raw transaction estimate for Q4 2022:  {q4_2022_raw:.2f}%")
print(f"Regression adjusted prediction:         {q4_2022_predicted:.2f}%")
print(f"Last reported quarter (2022Q3):         {comparison[comparison['quarter'] == '2022Q3']['reported_yoy'].values[0]:.2f}%")
```

```python
# calculate average sales and trend for each store using most recent 4 quarters
recent_sales = quarterly[quarterly['quarter'] >= '2022Q1']

# average quarterly sales per store
store_avg = recent_sales.groupby('locationid')['amount'].mean().reset_index()
store_avg.columns = ['locationid', 'avg_quarterly_sales']

# sales trend - compare last 2 quarters to first 2 quarters to see if store is improving or declining
early = recent_sales[recent_sales['quarter'] <= '2022Q2'].groupby('locationid')['amount'].mean()
late = recent_sales[recent_sales['quarter'] >= '2022Q3'].groupby('locationid')['amount'].mean()

trend = ((late - early) / early * 100).reset_index()
trend.columns = ['locationid', 'sales_trend_pct']

# combine into one table
store_metrics = store_avg.merge(trend, on='locationid', how='inner')

print(store_metrics.shape)

print("\n" + "="*50 + "\n")

print(store_metrics.describe())
```

```python
# check how many stores have extreme trend values
print(f"Stores with trend > 200%: {len(store_metrics[store_metrics['sales_trend_pct'] > 200])}")
print(f"Stores with trend < -80%: {len(store_metrics[store_metrics['sales_trend_pct'] < -80])}")

print("\n" + "="*50 + "\n")

# these extreme values are likely stores that were closed/opened mid period
# likely only have partial data in one of the two halves so data is not helpful
# cap the trend at the 5th and 95th percentile to remove noise
low = store_metrics['sales_trend_pct'].quantile(0.05)
high = store_metrics['sales_trend_pct'].quantile(0.95)

print(f"5th percentile:  {low:.1f}%")
print(f"95th percentile: {high:.1f}%")

store_metrics['sales_trend_pct'] = store_metrics['sales_trend_pct'].clip(low, high)

print("\n" + "="*50 + "\n")

print(store_metrics['sales_trend_pct'].describe())
```

```python
# join store metrics with webscrape location info
store_full = store_metrics.merge(webscrape, left_on='locationid', right_on='location_id', how='inner')

print(f"Stores with full data: {len(store_full)}")

print("\n" + "="*50 + "\n")

print(store_full[['locationid', 'city', 'state', 'avg_quarterly_sales', 'sales_trend_pct']].head(10))
```

```python
from scipy.spatial.distance import cdist
import numpy as np

# get lat/long for each store
coords = store_full[['locationid', 'latitude', 'longitude']].dropna()

# calculate distance between every pair of stores in miles
# first convert lat/long differences to miles (1 degree lat ~ 69 miles)
def haversine_miles(lat1, lon1, lat2, lon2):
    R = 3958.8  # radius of earth in miles
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# for each store find the distance to its nearest neighbor
min_distances = []
for _, row in coords.iterrows():
    distances = coords.apply(lambda x: haversine_miles(row['latitude'], row['longitude'], 
                                                        x['latitude'], x['longitude']), axis=1)
    # exclude the store itself (distance of 0)
    distances = distances[distances > 0]
    min_distances.append(distances.min())

coords['nearest_store_miles'] = min_distances

print(coords['nearest_store_miles'].describe())
```

```python
# calculate average sales by state as a proxy for market strength
state_sales = store_full.groupby('state')['avg_quarterly_sales'].mean().reset_index()
state_sales.columns = ['state', 'avg_state_sales']

# merge back onto store level data
store_full = store_full.merge(state_sales, on='state', how='left')

print(state_sales.sort_values('avg_state_sales'))

print("\n" + "="*50 + "\n")

print(f"Strongest state: {state_sales.loc[state_sales['avg_state_sales'].idxmax(), 'state']} - ${state_sales['avg_state_sales'].max():.0f}")
print(f"Weakest state:   {state_sales.loc[state_sales['avg_state_sales'].idxmin(), 'state']} - ${state_sales['avg_state_sales'].min():.0f}")
```

```python
# merge proximity data back onto store_full
store_full = store_full.merge(coords[['locationid', 'nearest_store_miles']], on='locationid', how='left')

# rank each factor from 0 to 1 where 0 = worst (most likely to close) and 1 = best
store_full['sales_rank'] = store_full['avg_quarterly_sales'].rank(pct=True)
store_full['trend_rank'] = store_full['sales_trend_pct'].rank(pct=True)
store_full['proximity_rank'] = 1 - store_full['nearest_store_miles'].rank(pct=True)  # closer = worse
store_full['market_rank'] = store_full['avg_state_sales'].rank(pct=True)

# combine into one closure score - equal weight to all four factors
store_full['closure_score'] = (store_full['sales_rank'] + 
                                store_full['trend_rank'] + 
                                store_full['proximity_rank'] + 
                                store_full['market_rank']) / 4

# sort by closure score to find weakest stores
weakest = store_full.sort_values('closure_score').head(20)

print(weakest[['city', 'state', 'avg_quarterly_sales', 'sales_trend_pct', 
               'nearest_store_miles', 'closure_score']].to_string())
```

```python
#display top 20
strongest = store_full.sort_values('closure_score').tail(20)

print(strongest[['city', 'state', 'avg_quarterly_sales', 'sales_trend_pct', 
               'nearest_store_miles', 'closure_score']].to_string())
```
